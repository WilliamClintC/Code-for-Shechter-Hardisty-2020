# Table 1 - Sample descriptives

Per-study: N, % male, mean age, mean self-reported competitiveness by gender.

Input: `data/1_derived/competition/wide_clean.csv` (the cleaned sample from Step 2).

Output: `output/2_analysis/tables/table1_descriptives.{csv,md}`.

```python
from pathlib import Path
import pandas as pd

ROOT = Path.cwd()
while not (ROOT / 'run_all.py').exists():
    if ROOT.parent == ROOT:
        raise RuntimeError('Could not find project root (run_all.py)')
    ROOT = ROOT.parent

TABLES = ROOT / 'output' / '2_analysis' / 'tables'
TABLES.mkdir(parents=True, exist_ok=True)

wide = pd.read_csv(ROOT / 'data' / '1_derived' / 'competition' / 'wide_clean.csv')
STUDY_LABELS = {'exp72': 'Study 1 (MTurk, 6-person)',
                'exp73': 'Study 2 (MTurk, 6-person)',
                'exp58': 'Study 3 (Undergrads, 10-person)'}
ORDER = ['exp72', 'exp73', 'exp58']
```

```python
# One row per study: N, % male, mean age, and mean self-reported competitiveness
# (`howcompetitive`, 1-7 scale) split by gender. Percentages and age are rounded
# to whole numbers, the competitiveness means to one decimal.
rows = []
for s in ORDER:
    sub = wide[wide['study'] == s]
    rows.append({
        'Study':            STUDY_LABELS[s],
        'N':                len(sub),
        '% male':           round((sub['gender'] == 'M').mean() * 100, 0),
        'Avg age':          round(sub['age'].mean(), 0),
        'Avg male comp':    round(sub.loc[sub['gender'] == 'M', 'howcompetitive'].mean(), 1),
        'Avg female comp':  round(sub.loc[sub['gender'] == 'F', 'howcompetitive'].mean(), 1),
    })
table1 = pd.DataFrame(rows)
print(table1.to_string(index=False))
```

```python
csv_path = TABLES / 'table1_descriptives.csv'
md_path  = TABLES / 'table1_descriptives.md'
table1.to_csv(csv_path, index=False)
with open(md_path, 'w') as f:
    f.write('# Table 1: Sample descriptives\n\n')
    f.write(table1.to_markdown(index=False))
    f.write('\n')
print(f'wrote {csv_path.relative_to(ROOT)}')
print(f'wrote {md_path.relative_to(ROOT)}')
```
